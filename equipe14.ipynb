```python
#Importação de bibliotecas
import pandas as pd
import plotly.express as px
```

```python
#Recebimento da tabela de dados
tabela = pd.read_csv('https://raw.githubusercontent.com/solon360/dataset-equipe-14/04cb7c6efc46bb6f151d680ad0b1175c7ad33a57/dados.csv')
```

```python
#Tratamento de dados

#Tabela em que os casos sem valores permanecem
erro = tabela["tipo_erro"]
tabela_nova = tabela
tabela_nova = tabela.drop(['falha', 'comp_subst'], axis=1)
tabela_nova["tipo_erro"] = tabela_nova["tipo_erro"].replace("error1",1)
tabela_nova["tipo_erro"] = tabela_nova["tipo_erro"].replace("error3",3)
tabela_nova["tipo_erro"] = tabela_nova["tipo_erro"].replace("error4",4)
tabela_nova["tipo_erro"] = tabela_nova["tipo_erro"].fillna(0)

#Tabela em que os casos sem valores foram excluídos
tabela_sem_vazios = tabela.drop(['tipo_erro','falha', 'comp_subst'], axis=1)
tabela_sem_vazios = tabela_sem_vazios.dropna(how = "any", axis = 0)
tabela_sem_vazios.insert(6, "tipo_erro", erro)
tabela_sem_vazios["tipo_erro"] = tabela_sem_vazios["tipo_erro"].replace("error1",1)
tabela_sem_vazios["tipo_erro"] = tabela_sem_vazios["tipo_erro"].replace("error3",3)
tabela_sem_vazios["tipo_erro"] = tabela_sem_vazios["tipo_erro"].replace("error4",4)
tabela_sem_vazios["tipo_erro"] = tabela_sem_vazios["tipo_erro"].fillna(0)
```

```python
print("Medianas Tabela A")
print(tabela_nova["volt"].median())
print(tabela_nova["pressao"].median())
print(tabela_nova["vibracao"].median())
print(tabela_nova["rotacao"].median())

print("\nMedianas Tabela B")
print(tabela_sem_vazios["volt"].median())
print(tabela_sem_vazios["pressao"].median())
print(tabela_sem_vazios["vibracao"].median())
print(tabela_sem_vazios["rotacao"].median())
```
